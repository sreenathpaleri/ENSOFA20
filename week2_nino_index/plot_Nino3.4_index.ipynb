```python
df = pd.read_csv("nino34.csv")
del df['Column1']
df = df.set_index('Column2')
df.columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df.index.names = ['Year']
df=df.drop([1948, 1949])
df['Dec'].iloc[-1]=None
df['Nov'].iloc[-1]=None
df['Oct'].iloc[-1]=None
```

```python
#EP
nsvn = df.loc[1997]
neit = df.loc[1998]

#CP
ofour = df.loc[2004]
ofive = df.loc[2005]

#normal
one = df.loc[1993]
two = df.loc[1994]
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, gridspec_kw={'hspace': .5, 'wspace': .025})
f.suptitle("Nino 3.4 Index", size=15)
f.set_figheight(5)
f.set_figwidth(15)
ax1.plot(nsvn)
ax1.set_ylim(24.8, 29.3)
ax1.set_title('1997')
ax2.plot(neit)
ax2.set_title('1998')
```
